```python
from langchain_community.document_loaders import JSONLoader
import json
from pprint import pprint
import os
from langchain.text_splitter import CharacterTextSplitter
from langchain.schema.document import Document
import warnings
warnings.filterwarnings('ignore')
```

```python
def get_text_chunks_langchain(text,meta):
    text_splitter = CharacterTextSplitter(chunk_size=2056, chunk_overlap=128)
    docs = [Document(page_content=x) for x in text_splitter.split_text(text)]
    for doc in docs:
        doc.metadata = meta
    return docs

def getMeta(d):
    if "fullText" in d:
        del d["fullText"]
        del d["Summary"]
        d['authors'] = "Wikipedia"
        d['publication_date'] = "2024"
    return d
```

```python
path = 'wiki/'
filenames = os.listdir(path)

documents = []
for file in filenames:
    data = json.loads(Path(path+file).read_text())
    docs = get_text_chunks_langchain(data['fullText'],getMeta(data))
    documents.append(docs)
    
```
